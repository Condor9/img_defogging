```python
# 

import cv2
import matplotlib.pyplot as plt
import numpy as np

original = cv2.imread(r'./lena.png', 0)  # 

dft = np.fft.fft2(original)
dftShift = np.fft.fftshift(dft)     # 
result = 20*np.log(np.abs(dftShift))

plt.subplot(121), plt.imshow(original, cmap='gray')
plt.title('original')
plt.axis('off')

plt.subplot(122), plt.imshow(result, cmap='gray')
plt.title('fft')
plt.axis('off')

plt.show()
```

```python
# 

import cv2
import matplotlib.pyplot as plt
import numpy as np

original = cv2.imread(r'./lena.png', 0)  # 

dft = np.fft.fft2(original)
dftShift = np.fft.fftshift(dft)     # 
result = 20*np.log(np.abs(dftShift))

ishift = np.fft.ifftshift(dftShift)     # 
iImg = np.fft.ifft2(ishift)                 # 
iImg = np.abs(iImg)

plt.subplot(131), plt.imshow(original, cmap='gray')
plt.title('original')
plt.axis('off')

plt.subplot(132), plt.imshow(result, cmap='gray')
plt.title('fft')
plt.axis('off')

plt.subplot(133), plt.imshow(iImg, cmap='gray')
plt.title('infft')
plt.axis('off')
plt.show()
```
