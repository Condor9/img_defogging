```python
from torchvision import transforms
import matplotlib.pyplot as plt
from PIL import Image
```

```python
ori_pic=Image.open('njust.jpg')
display(ori_pic)
```

```python
#平移
trans=transforms.RandomAffine(degrees=0,translate=(0.4,0.2))(ori_pic)
#旋转
rot=transforms.RandomAffine(degrees=(22,38))(ori_pic)
#水平翻转
flip1=transforms.RandomHorizontalFlip(p=1)(ori_pic)
#垂直翻转
flip2=transforms.RandomVerticalFlip(p=1)(ori_pic)
```

```python
plt.xticks([])
plt.yticks([])
plt.figure(1); plt.xlabel("origin"); plt.imshow(ori_pic)

plt.xticks([])
plt.yticks([])
plt.figure(2); plt.xlabel("translation"); plt.imshow(trans)

plt.xticks([])
plt.yticks([])
plt.figure(3); plt.xlabel("rotation"); plt.imshow(rot)

plt.xticks([])
plt.yticks([])
plt.figure(4); plt.xlabel("horizontalflip"); plt.imshow(flip1)

plt.xticks([])
plt.yticks([])
plt.figure(5); plt.xlabel("verticalflip"); plt.imshow(flip2)
```
